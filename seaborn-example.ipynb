```python
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
ax = sns.countplot(x="class", data=titanic, palette="Blues", hue='sex')
```

```python
ax = sns.histplot(data=titanic, x="age", kde=True, bins=5, hue="sex")
```

```python
ax = sns.boxplot(data=titanic, x="class", y="age", hue="sex", palette="Greens")
```

```python
ax = sns.scatterplot(data=titanic, x="age", y="fare", hue="alive", palette="Greens")
```

```python
ax = sns.jointplot(data=titanic, x="age", y="fare", hue="alive", palette="Greens")
```

```python
ax = sns.jointplot(data=titanic, x="age", y="fare", palette="Greens", kind="reg")
```
